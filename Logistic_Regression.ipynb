###***IMPORTING LIBRARIES***

```python
import numpy as np
import pandas as pd
```

###***LOADING THE DATASET***

```python
data = pd.read_csv("diabetes2.csv")
data
```

###***SUMMARIZING THE DATASET***

```python
print(data.shape)
print(data.head())
```

###***SEGREGATING THE DATASET INTO INPUT AND OUTPUT***

```python
input  = data.iloc[:,:-1].values
print(input)

output = data.iloc[:,-1:].values
print(output)
```

###***FEATURE SCALING***

```python
from sklearn.preprocessing import StandardScaler
scaling      = StandardScaler()
scaled_input = scaling.fit_transform(input)

scaled_input
```

###***SPLITTING THE DATASET INTO TRAIN AND TEST DATA***

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(scaled_input, output,test_size=0.20)
```

###***TRAINING THE DATASET***

```python
from sklearn.linear_model import LogisticRegression

Logistic_Regression_Model = LogisticRegression()

Logistic_Regression_Model.fit(x_train,y_train)
```

###***PREDICTING THE RESULT USING THE TRAINED MODEL***

```python
y_pred = Logistic_Regression_Model.predict(x_test)
y_pred
```

###***CALCULATING THE ACCURACY OF THE TRAINED MODEL***

```python
from sklearn.metrics import confusion_matrix, accuracy_score
print("confusion_matrix:")
print(confusion_matrix(y_test,y_pred))
print("accuracy_score:",accuracy_score(y_test,y_pred)*100,"%")
```
